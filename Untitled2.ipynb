```python
import pandas as pd

# CSV dosyanı Python'a bir veri çerçevesi (DataFrame) olarak tanıtıyoruz
# Dosya adını kendi kaydettiğin adla değiştir
df = pd.read_csv('olist_sellers_dataset.csv')

# Verinin ilk 5 satırını kontrol edelim
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. En çok satıcıya sahip ilk 10 şehri hesaplayalım
top_cities = df['seller_city'].value_counts().head(10).reset_index()
top_cities.columns = ['Sehir', 'Satici_Sayisi']

# 2. Grafik boyutunu ve temasını ayarlayalım
plt.figure(figsize=(12, 6))
sns.set_theme(style="whitegrid")

# 3. Yatay bar grafiği çizelim
sns.barplot(data=top_cities, x='Satici_Sayisi', y='Sehir', palette='viridis')

# 4. Grafiğe başlık ve eksen isimleri ekleyelim
plt.title('Olist Platformundaki Satıcıların En Çok Bulunduğu İlk 10 Şehir', fontsize=14, fontweight='bold')
plt.xlabel('Toplam Satıcı Sayısı', fontsize=12)
plt.ylabel('Şehirler', fontsize=12)

# 5. Grafiği ekranda gösterelim
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Tabloları okuyalım
orders = pd.read_csv('olist_orders_dataset.csv')
items = pd.read_csv('olist_order_items_dataset.csv')

# Sipariş tarihi ve fiyat bilgisini sipariş ID'sine göre birleştirelim
merged_orders = pd.merge(orders, items, on='order_id', how='inner')

# Tarih sütununu Python'ın anlayacağı zaman formatına çevirelim
merged_orders['order_purchase_timestamp'] = pd.to_datetime(merged_orders['order_purchase_timestamp'])

# Yıl-Ay bazında gruplayıp toplam ciroyu (price) hesaplayalım
merged_orders['year_month'] = merged_orders['order_purchase_timestamp'].dt.to_period('M')
monthly_sales = merged_orders.groupby('year_month')['price'].sum().reset_index()

# Grafik için tarihi string formatına geri getirelim ve son eksik ayı (varsa) temizleyelim
monthly_sales['year_month'] = monthly_sales['year_month'].astype(str)
monthly_sales = monthly_sales.sort_values('year_month').iloc[:-1] # Son ayı tam tamamlanmadığı için kırpıyoruz

# Grafiği çizelim
plt.figure(figsize=(14, 6))
sns.lineplot(data=monthly_sales, x='year_month', y='price', marker='o', color='#1f77b4', linewidth=2.5)

plt.title('Olist E-Ticaret Aylık Ciro Gelişimi (Zaman Serisi)', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Zaman (Yıl-Ay)', fontsize=12)
plt.ylabel('Toplam Ciro (R$ - Brezilya Reali)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# Tabloları okuyalım
orders = pd.read_csv('olist_orders_dataset.csv')
reviews = pd.read_csv('olist_order_reviews_dataset.csv')

# Sadece teslim edilmiş siparişleri alalım ve tarihleri formatlayalım
delivered_orders = orders[orders['order_status'] == 'delivered'].copy()
delivered_orders['order_purchase_timestamp'] = pd.to_datetime(delivered_orders['order_purchase_timestamp'])
delivered_orders['order_delivered_customer_date'] = pd.to_datetime(delivered_orders['order_delivered_customer_date'])

# Teslimat süresini gün cinsinden hesaplayalım (Teslim alınan tarih - Sipariş tarihi)
delivered_orders['delivery_time_days'] = (delivered_orders['order_delivered_customer_date'] - delivered_orders['order_purchase_timestamp']).dt.days

# Yorum tablosuyla birleştirelim
delivery_review = pd.merge(delivered_orders, reviews, on='order_id', how='inner')

# Anlamlı bir grafik için teslimat süresi 0 ile 50 gün arasında olan normal siparişlere odaklanalım (Aykırı değerleri elemek için)
filtered_data = delivery_review[(delivery_review['delivery_time_days'] >= 0) & (delivery_review['delivery_time_days'] <= 50)]

# Kutu Grafiği (Boxplot) Çizimi
plt.figure(figsize=(12, 6))
sns.boxplot(data=filtered_data, x='review_score', y='delivery_time_days', hue='review_score', palette='autumn', legend=False)

plt.title('Teslimat Süresinin Müşteri Memnuniyet Puanına Etkisi', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Müşteri Memnuniyet Puanı (1-5 Yıldız)', fontsize=12)
plt.ylabel('Teslimat Süresi (Gün)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
